```python
#read dataset
import pandas as pd 
students=pd.read_csv("https://modcom.co.ke/data/datasets/schoolcleaned.csv")  
students.head()    
```

```python
#visualisation libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use("seaborn")
```

```python
#piechart
students.groupby('Gender').size().plot(kind='pie', autopct = '%2.1f%%')
plt.title('Distribution of Gender')
```

```python
students.groupby(['HowCommute', 'State']).size().plot(kind='pie',autopct='%2.1f%%') 
plt.title('Distribution of how students commute by state')
```

```python
#show the mean of math for students based on their gender and state then visualise the results on a bar graph
students.groupby(['Gender', 'State'])['Math'].mean().plot(kind='bar')
```

```python
#count plot -> depict the number of count on a categorical data
ax = sns.countplot(students['State'])
for p in ax.patches:
  txt =str(p.get_height().round(1))+''
  txt_x = p.get_x()
  txt_y = p.get_height()
  ax.text(txt_x, txt_y, txt)
```

```python
#distribtion of contionous variable
sns.distplot(students['Height'], kde = False)
```

```python
#scatter plot
#shows two continous values and how they are related

sns.scatterplot(x='Weight',y='Height', data=students)
```

```python
#heatmap-> depicts relationship bettween continous variable
subset = students[['Weight', 'Height','Reading', 'Writing','Sprint']]
sns.heatmap(subset.corr(), annot=True, cmap='Greens')
```

```python
#botplot ->generates the outliers (unusual data plots)
sns.boxplot(y='Math', x='Rank', data = students)
#percentiles
#upper percentile Q3
#lower percentile(25%) Q1
#IQR= Q3-Q1
#outliers_upper = Q3 + IQR *1.5
#outliers_lower = Q1 + IQR *1.5
```

```python
#Concentration of the data point
#violin plot
sns.violinplot(x='Gender',y='Reading', data= students)
```

```python

#swarmplot
sns.swarmplot(x='Gender',y='Reading', data= students)
```
