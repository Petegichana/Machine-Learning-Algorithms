```python
import pandas as pd
data=pd.read_csv('https://modcom.co.ke/data/datasets/pima.csv')
data
```

```python
data.isnull().sum()
```

```python
array = data.values
X = array[:,0:8]
Y = array[:, 8]
```

```python
#model selection
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size=0.30, random_state=42)
```

```python
#algorithms
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
```

```python
#cross validation
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, Y_train)
```

```python
#testing 
predictions = model.predict(X_test)
print('We have predicted', predictions)
```

```python
print('The accurate answers are', Y_test)
```

```python
#accuracy estimation
from sklearn.metrics import accuracy_score
print('ACCURACY: ', accuracy_score(predictions, Y_test))
```

```python
#PREDITIONS

symptoms = [[2,100,50,30,1,20,0.50,28]]
disease = model.predict(symptoms)
print('You are 74% likely to be', disease)
```
